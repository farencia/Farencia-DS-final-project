```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from matplotlib.ticker import FuncFormatter  # Import formatter

df = pd.read_csv("artDataset2.csv")

# Defining the top 10 artists
top_artists = ["Russell Young", "Alex Katz", "Ed Ruscha", "Cindy Sherman", 
               "Vija Celmins", "Frank Schroeder", "Gustave Blache III", 
               "Biff Elrod", "Jasper Johns", "Larry Bell"]

# Filter dataset for these artists
filtered_df = df[df["artist"].isin(top_artists)]

# Drop rows where "art movement" is missing
filtered_df = filtered_df.dropna(subset=["art movement"])

# Count occurrences of each artist per art movement
artists_per_movement = filtered_df.groupby(["art movement", "artist"]).size().reset_index(name="count")

# Count unique artists per art movement
movement_counts = artists_per_movement.groupby("art movement")["artist"].nunique().reset_index()

# Calculate the percentage of top 10 artists associated with each movement
total_top_artists = len(top_artists)
movement_counts["percentage"] = (movement_counts["artist"] / total_top_artists) * 100

# Get unique art movements
art_movement_names = movement_counts["art movement"].tolist()

# Prepare data for the boxplot
# Since we don't have distributions, we simulate variation by using each movement's percentage multiple times
boxplot_data = [np.random.normal(perc, 2, size=10) for perc in movement_counts["percentage"]]

# Seaborn color palette (Set2) for colors
colors = sns.color_palette("Set2", n_colors=len(art_movement_names))

# Create the box plot
fig, ax = plt.subplots(figsize=(10, 6))
ax.set_ylabel("Percentage of top 10 artists")
ax.set_title("Top 10 artists across art movements", fontsize=15, fontweight='bold')
ax.set_xlabel("Art movements covered by the top 10 artists")

bplot = ax.boxplot(boxplot_data, patch_artist=True, labels=art_movement_names)
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Fill boxes with colors
for patch, color in zip(bplot['boxes'], colors):
    patch.set_facecolor(color) # setting the box colors

# Function to format y-axis labels as percentages
def percent_formatter(x, pos):
    return f"{x:.0f}%"

# Apply the formatter to the y-axis
ax.yaxis.set_major_formatter(FuncFormatter(percent_formatter))

plt.show()
plt.tight_layout()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from ipywidgets import interact, widgets

# Load dataset
df = pd.read_csv("artDataset2.csv")

# Converting the year column to numeric 
df["year created"] = pd.to_numeric(df["year created"], errors="coerce")

# Get unique art movements
art_movements = df["art movement"].unique()

# Set Seaborn color palette
set2_colors = sns.color_palette("Set2")

# Define interactive function
def plot_art_prices(selected_movement):
    # Filter dataset based on selected movement
    filtered_df = df[df["art movement"] == selected_movement]
    
    # Sort by year for proper visualization
    filtered_df = filtered_df.sort_values(by="year created")
    
    # Plot with a color from the Set2 palette
    plt.figure(figsize=(12, 6))
    sns.lineplot(x="year created", y="price(USD)", data=filtered_df, marker="o", linewidth=2.5, color=set2_colors[0])  #

    
    # Styling
    plt.xlabel("year created", fontsize=14)
    plt.ylabel("price(USD)", fontsize=14)
    plt.title(f"Price Trends for {selected_movement}", fontsize=16, fontweight="bold")
    plt.grid(True, linestyle="--", alpha=0.6)
    
    # Show plot
    plt.show()

# Making the  dropdown widget
interact(plot_art_prices, selected_movement=widgets.Dropdown(options=art_movements, description="Art Movement:"));
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.ticker as mticker
import seaborn as sns

df = pd.read_csv("artDataset2.csv")

# Grouping by year and calculate the average price
avg_price_by_year = df.groupby("year created")["price(USD)"].mean()
avg_price_by_period = df.groupby("time period")["price(USD)"].mean()
# Grouping by time period and calculate the average price, sorted in descending order

set2_colors = sns.color_palette("Set2", n_colors=len(avg_price_by_period))
# Importing the Pastel Seaborn color palette

# creating  a list of values for all periods over time
years = avg_price_by_year.index
period_data = []
for period in avg_price_by_period.index:
    # Get average price for each period per year
    period_avg_price = df[df["time period"] == period].groupby("year created")["price(USD)"].mean()
    period_data.append(period_avg_price.reindex(years, fill_value=0))  # Fill missing years with 0

# Convert period_data into a NumPy array for stackplot
period_data = np.array(period_data)

# Create the plot
fig, ax = plt.subplots(figsize=(10, 6))

# Stackplot for each time period's average price over the years
ax.stackplot(years, period_data, labels=avg_price_by_period.index, colors=set2_colors, alpha=0.8)

# Adding the legend and labels as guides 
ax.legend(loc='upper left', reverse=True)
ax.set_title('Average artwork price by time period over time', fontweight="bold")
ax.set_xlabel('Year',fontweight="bold" )
ax.set_ylabel('Average price (USD)', fontweight="bold")

# Set the y-axis ticks to be more readable
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, loc: "${:,}".format(int(x))))


plt.tight_layout()
plt.show()

df = pd.DataFrame(period_data)
print(df)
```
